```python
def sep_yearmonths(x):
    x['Date'] = x['Date'][:-2]
    return x
cov19 = cov19.apply(sep_yearmonths,axis=1)
```

```python
cov19["Monthly Tests"] = cov19.groupby(['Entity','Date']).sum("Daily tests")
cov19
```

```python
cov19['Date']=cov19['Date'].str.replace("-","", regex=False).astype(float)
cov19['Date']=pd.to_datetime(cov19['Date'],format="%Y%m%d")
```

```python
cur.execute("CREATE TABLE cov19(
            Entity, Continent, Latitude, Longitude, Avg Temp per year, "Hosp beds per 1000", "Doctors per 1000" , 
            "GDP/capita", Population, Median Age, "Age > 65", Date, Daily tests, Cases, Deaths)
            ")
for row in cov19.itertuples():
    cursor.execute('''
                INSERT INTO products (Entity, Continent, Latitude, Longitude, Avg Temp per year, "Hosp beds per 1000", "Doctors per 1000" , 
                "GDP/capita", Population, Median Age, "Age > 65", Date, Daily tests, Cases, Deaths)
                VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
                ''',
                row.product_id, 
                row.product_name,
                row.price
                )
conn.commit()
```
